### Gathering Data (Part 2)

To obtain the necessary values to test my heuristics, I need to query user information from the Twitter API again.

```python
import numpy as np
import pandas as pd
import requests
import time
```

```python
sample_df = pd.read_pickle("data/processed_data/sample_data_with_usernames.pkl")
```

```python
def searchTwitterUsers(usernames):
    json_stuff = []
    for un in usernames:
        un_str = ",".join(usernames)
        users = f"usernames={un_str}"
        user_fields = "user.fields=username,description,created_at,profile_image_url,public_metrics,verified"
        url = "https://api.twitter.com/2/users/by?{}&{}".format(users, user_fields)
        bearer_token = ...
        headers = {"Authorization": f"Bearer {bearer_token}"}

        response = requests.request("GET", url, headers=headers)
        if response.status_code != 200:
            print("Request returned an error: {} {}".format(response.status_code, response.text))
            if response.status_code == 429:
                print("Sleeping for 900 seconds")
                time.sleep(900)
                j = response.json()
            else:
                if un is None:
                    j = None
        else:
            j = response.json()
        json_stuff.append(j)
        break
    return json_stuff
```

```python
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
usernames_with_partitions = list(chunks(list(sample_df["usernames"].value_counts().keys()), 10))
```

```python
json_data = []
```

```python
for i in range(len(usernames_with_partitions)):
    unwp = usernames_with_partitions[i]
    json_data.append(searchTwitterUsers(unwp))
    print(f"{i} completed...")
```

```python
df_list = [json_data[i][0]["data"] for i in range(len(json_data))]
```

```python
user_df = pd.concat([pd.DataFrame(i) for i in df_list], axis=0, ignore_index=True)
```

```python
sample_df['usernames'][sample_df['usernames'].isnull()] = ""
```

```python
print(list(user_df["public_metrics"])[0])
```

```python
sample_df["user_created_at"] = ["None"] * len(sample_df)
sample_df["user_description"] = ["None"] * len(sample_df)
sample_df["user_profile_image_url"] = ["None"] * len(sample_df)
sample_df["user_followers"] = ["None"] * len(sample_df)
sample_df["user_following"] = ["None"] * len(sample_df)
sample_df["user_tweet_count"] = ["None"] * len(sample_df)
```

```python
for i in range(len(user_df["username"])):
    username = list(user_df["username"])[i]
    print(username)
    sample_df["user_created_at"][sample_df["usernames"] == username] = list(user_df["created_at"])[i]
    sample_df["user_description"][sample_df["usernames"] == username] = list(user_df["description"])[i]
    sample_df["user_profile_image_url"][sample_df["usernames"] == username] = list(user_df["profile_image_url"])[i]
    sample_df["user_followers"][sample_df["usernames"] == username] = list(user_df["public_metrics"])[i]["followers_count"]
    sample_df["user_following"][sample_df["usernames"] == username] = list(user_df["public_metrics"])[i]["following_count"]
    sample_df["user_tweet_count"][sample_df["usernames"] == username] = list(user_df["public_metrics"])[i]["tweet_count"]
    
```
